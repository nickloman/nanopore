__poretools__ can run either on an individual FAST5 file, a directory containing FAST5 files, or a tar archive of FAST5 files.

Here we set up several variables which relate to the locations of the NONI-R7, ONI-R7-1, ONI-R7-2 and R7.3 files.

```python
noni="/dev/shm/flowcells/NONI"
oni1="/dev/shm/flowcells/flowcell_17/downloads"
oni2="/dev/shm/flowcells/flowcell_18/downloads"
r73 ="/dev/shm/flowcells/flowcell_20/downloads"
```

__poretools__ has a number of different command line options. Running __poretools__ with no parameters gives us a brief list (and complies with [Torsten's first rule](http://www.gigasciencejournal.com/content/2/1/15))

```python
!poretools
```

We can get more information if we run __poretools__ with the -h (help) option.

```python
!poretools -h
```

Let's start with a simple one, the ``stats`` command, this will give us some basic statistics about our reads.

The ``-q`` option stops ``poretools`` outputting any warning messages.

```python
!poretools stats -q --full-tsv $noni > noni.stats.txt
!poretools stats -q --full-tsv $oni1 > oni1.stats.txt
!poretools stats -q --full-tsv $oni2 > oni2.stats.txt
!poretools stats -q --full-tsv $r73 > r73.stats.txt
```

```python
!grep ^ noni.stats.txt oni1.stats.txt oni2.stats.txt r73.stats.txt
```

```python
!grep ^ noni.stats.txt oni1.stats.txt oni2.stats.txt r73.stats.txt | sed 's/:/\t/' > runstats.txt
```

```python
%%R
library(ggplot2)
library(reshape2)
runstats = read.table("runstats.txt", header=T, sep="\t")
runstats$File = sub("noni.stats.txt", "R7 Non-overnight", runstats$File)
runstats$File = sub("oni1.stats.txt", "R7 Overnight 1", runstats$File)
runstats$File = sub("oni2.stats.txt", "R7 Overnight 2", runstats$File)
runstats$File = sub("r73.stats.txt", "R7.3", runstats$File)
runstats$Cat = sub("template", "Template", runstats$Cat)
runstats$Cat = sub("complement", "Complement", runstats$Cat)
runstats$Cat = sub("twodirections", "Normal 2D", runstats$Cat)
runstats$Cat = sub("2D_hq", "Full 2D", runstats$Cat)
runstats_casted = dcast(runstats, ... ~ Cat)
runstats_melted = melt(runstats_casted, id.vars=c('File', 'Subcat', 'files') )
runstats_melted[1,]
runstats_melted$newvar = factor(runstats_melted$variable, levels=c('Template', 'Complement', 'Normal 2D', 'Full 2D'))
ggplot(subset(runstats_melted, Subcat == 'total reads'), aes(x=File, y=value/files, fill=newvar)) + geom_bar(stat="identity", position="dodge") + facet_wrap ( ~ Subcat, scales="free_y" ) + theme_bw(base_size=22) + scale_x_discrete("Run") + scale_fill_discrete(guide=guide_legend(title="Read type")) + ylim(0,1)+ scale_y_continuous("Proportion") 
ggsave("Yield_proportion.pdf", width=12, height=8)
```
